## PySimpleGUI Bread and Butter Code:

```python
### REFERENCE MATERIAL FOR PySimpleGUI: https://www.pysimplegui.org/en/latest/
### This cell is the bread and butter/ foundation that allows PySimpleGUI to create 
### a simple yet nice entry table for users to key in the values (Try running the cell!!)
### Dictionary Tutorial Reference: https://www.w3schools.com/python/python_dictionaries_access.asp (IMPORTANT FOR PySimpleGUI!!!!!)
import PySimpleGUI as sg


sg.theme('DarkTeal9')

layout = [  [sg.Text('Some text on Row 1')],
          [sg.Text('Enter something on Row 2'), sg.InputText(key = 'something')], ## Not defining anything inside sg.InputText() defaults the value index as [0], if we specify a key, the value will be stored as the new key
          [sg.Text('Enter the name of your crush'), sg.InputText(key = 'Crush')],
          [sg.Button('Ok'), sg.Button('Cancel')] ]

window = sg.Window('Window Title', layout)
# Event Loop to process "events" and get the "values" of the inputs
while True:
    event, values = window.read()
    if event == sg.WIN_CLOSED or event == 'Cancel': # if user closes window or clicks cancel
        break
    print('You entered ', values)

window.close()
```

```python
import PySimpleGUI as sg
import pandas as pd
import datetime 

sg.theme('DarkAmber')

##You can clear the selected value of a Combobox by setting its value to an empty string:
# ComboBox.set('')


## Functions that performs: 
# 1) Selected value clearing
# 2) Form validation (Ensures all inputs are key in properly. NO BLANKS)

def clear_input(window): ### This one is for the Clear button. Clear input function clears all values, including the title of my buttons lmao. So i have to specify more detailed. REFEREMCE: https://stackoverflow.com/questions/72973496/clearing-date-picker-in-pysimplegui
    for key, values in window.key_dict.items():
        if isinstance(values, sg.Combo) or isinstance(values, sg.Input): ## This if loop helps to clear both combobox and input values, successfully ignoring the Button titles. Cant believe this worked.
            values.update(value='')
            
# def clear_input(): ## This one is for the submit button, cause the previous clear input function only clears Input values, combo box values wont get cleared!!!
#     for key in values:
#         window[key]('')
#     return None

Excel = 'C:/My Files/Python_Projects/Python Tutorials/Python Stuff/On-Going Projects/Data Entry GUI/Cartridge_Tracking.xlsx'

df = pd.read_excel(Excel)

Current_Time = datetime.datetime.now() ## I use this line to get the current date and time

sg.set_options(font = 12)

layout = [
          [sg.Text('Please fill Up All Details Accordingly!', text_color = 'red', font = 15)], ## Font changes the size of the font
          [sg.Text('Please input current Date: ', size = (30,1)), sg.InputText(key = 'Date'), sg.CalendarButton('Submission Date', close_when_date_chosen = True, target = 'Date', size= (20,1))], ## Size adjusts the size of the text
          [sg.Text('Please input WWID: ',size = (30,1)), sg.InputText(key = 'WWID')], 
            
          [sg.Text('Link: ', size = (30,1)), sg.InputText(key = 'Link')],
          [sg.Text('Cartridge ID Returned', size = (30,1)), sg.InputText(key = 'Cartridge_Returned',)],  ## Not defining anything inside sg.InputText() defaults the value index as [0], if we specify a key, the value will be stored as the new key         
         ## Since I can select more than one checkbox, consider radio buttons/Combobox. REFERENCE: https://holypython.com/gui-with-python-checkboxes-and-radio-buttons-pysimplegui-part-ii/
          [sg.Text('Returned Cartridge Type', size = (30,1)), sg.Combo(['RS2-9F', 'RS1-9A', 'RS1-6F', 'RT4-6F' ], key = 'Returned Cartridge Type')], 

         [sg.Text('Cartridge ID Taken', size = (30,1)), sg.InputText(key = 'Cartridge_Taken')],                                 
         [sg.Text('Taken Cartridge Type', size = (30,1)), sg.Combo(['RS2-9F', 'RS1-9A', 'RS1-6F', 'RT4-6F'], key = 'Taken Cartridge Type')], 
                                           
          
          [sg.Text('Reason of Return: ', size = (30,1)), sg.Combo(['New Setup','OOC','EOD/Splashing'], key = 'Reason')],
          
          [sg.Button('Submit'),sg.Button('Clear'), sg.Button('Cancel')] 

          ]

window = sg.Window('Cartridge Tracking GUI', layout)

input_key_list = [key for key, value in window.key_dict.items() if isinstance (value, sg.Combo) or isinstance (value, sg.Input)]## Python to check if all input values are placed: Reference: https://stackoverflow.com/questions/68602498/check-if-all-inputs-has-values-in-pysimplegui
 ## REFERENCE FOR Isinstance https://www.w3schools.com/python/ref_func_isinstance.asp#:~:text=The%20isinstance()%20function%20returns,the%20types%20in%20the%20tuple.
# Event Loop to process "events" and get the "values" of the inputs

while True:
    
    event, values = window.read()
    if event == sg.WIN_CLOSED or event == 'Cancel': # if user closes window or clicks cancel
        break
    if event == 'Clear':
        clear_input(window)
    if event == 'Submit': # If user clicks Submit
        if all(map(str.strip, [values[key] for key in input_key_list])): ## Same reference used as the input_key_list.

            df.to_excel(Excel, index = False)
            sg.popup('Data Saved!!')
            clear_input(window)
            df = df.append(values, ignore_index = True) ## NOTE: Can only append a dict if ignore_index=True
            
### Python to check if all input values are placed: 
##Reference: 
#https://stackoverflow.com/questions/68602498/check-if-all-inputs-has-values-in-pysimplegui AND 
#https://www.w3schools.com/python/ref_func_all.asp AND 
#https://www.w3schools.com/python/ref_func_map.asp
#https://www.youtube.com/watch?v=gJLasS_-NX0&list=PLqK_fRVXlXeY89mOYRYGElWbGSwIMoDxC&index=6
            
        else:
            sg.popup('Form not completely filled up!')
        
window.close()
```

```python
import PySimpleGUI as sg
import pandas as pd
import datetime 

sg.theme('DarkTeal9')

def clear_input():
    for key in values:
        window[key]('')
    return None

Excel = 'C:/My Files/Python_Projects/Python Tutorials/Python Stuff/On-Going Projects/Data Entry GUI/Cartridge_Tracking.xlsx'

df = pd.read_excel(Excel)

Current_Time = datetime.datetime.now() ## I use this line to get the current date and time

layout = [[sg.Text('Please fill Up All Details Accordingly!', text_color = 'red', font = 12)], ## Font changes the size of the font
          [sg.Text('Please input current Work Week (WW): ', size = (30,1)), sg.InputText(key = 'Work_Week')], ## Size adjusts the size of the text
          [sg.Text('Please input WWID: ',size = (30,1)), sg.InputText(key = 'WWID')], 
          [sg.Text('Cartridge ID Returned', size = (30,1)), sg.InputText(key = 'Cartridge_Returned',)],  ## Not defining anything inside sg.InputText() defaults the value index as [0], if we specify a key, the value will be stored as the new key         
          
          
          ## Since I can select more than one checkbox, consider radio buttons. REFERENCE: https://holypython.com/gui-with-python-checkboxes-and-radio-buttons-pysimplegui-part-ii/
          [sg.Text('Returned Cartridge Type', size = (30,1)), 
                                           sg.Radio('RS2-9F','Cartridge R',key = 'Returned Cartridge Type'),
                                           sg.Radio('RS1-6F','Cartridge R',key = 'Returned Cartridge Type'),
                                           sg.Radio('RS1-9A','Cartridge R',key = 'Returned Cartridge Type'),
                                           sg.Radio('RT4-6F','Cartridge R',key = 'Returned Cartridge Type'),
                                           sg.Radio('Others: ','Cartridge R'), sg.InputText(key = 'Returned Cartridge Type', size = 10)],
         [sg.Text('Cartridge ID Taken', size = (30,1)), sg.InputText(key = 'Cartridge_Taken')],                                 
         [sg.Text('Taken Cartridge Type', size = (30,1)), 
                                           sg.Radio('RS2-9F','Cartridge T',key = 'Taken Cartridge Type'),
                                           sg.Radio('RS1-6F','Cartridge T',key = 'Taken Cartridge Type'),
                                           sg.Radio('RS1-9A','Cartridge T',key = 'Taken Cartridge Type'),
                                           sg.Radio('RT4-6F','Cartridge T',key = 'Taken Cartridge Type'),
                                           sg.Radio('Others: ','Cartridge T'), sg.InputText(key = 'Taken Cartridge Type', size = 10)],   
          
          [sg.Text('Reason of Return: ', size = (30,1)), sg.Combo(['New Setup','OOC','EOD/Splashing'], key = 'Reason')],
          
          [sg.Button('Submit'),sg.Button('Clear'), sg.Button('Cancel')] ]

window = sg.Window('Cartridge Tracking GUI', layout)

# Event Loop to process "events" and get the "values" of the inputs
while True:
    event, values = window.read()
    if event == sg.WIN_CLOSED or event == 'Cancel': # if user closes window or clicks cancel
        break
    if event == 'Clear':
        clear_input()
    if event == 'Submit': # If user clicks Submit
        df = df.append(values, ignore_index = True) ## NOTE: Can only append a dict if ignore_index=True
        df.to_excel(Excel, index = False)
        sg.popup('Data Saved!!')
        clear_input()
        
window.close()
```
